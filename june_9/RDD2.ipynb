```python
!pip install pyspark

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("OnlineLearningPlatform") \
        .master("local[*]") \
            .getOrCreate()

sc = spark.sparkContext
```

```python
transactions = [
  "DEPOSIT","WITHDRAW","TRANSFER","DEPOSIT","BALANCE_CHECK",
  "TRANSFER","DEPOSIT","WITHDRAW","LOAN_PAYMENT","DEPOSIT",
  "TRANSFER","WITHDRAW","DEPOSIT","BALANCE_CHECK","TRANSFER",
  "DEPOSIT","WITHDRAW","LOAN_PAYMENT","TRANSFER","DEPOSIT",
  "WITHDRAW","DEPOSIT","TRANSFER","BALANCE_CHECK","WITHDRAW",
  "DEPOSIT","TRANSFER","LOAN_PAYMENT","DEPOSIT","WITHDRAW",
  "TRANSFER","DEPOSIT","BALANCE_CHECK","TRANSFER","WITHDRAW",
  "DEPOSIT","TRANSFER","LOAN_PAYMENT","DEPOSIT","WITHDRAW",
  "TRANSFER","BALANCE_CHECK","DEPOSIT","TRANSFER","WITHDRAW",
  "LOAN_PAYMENT","DEPOSIT","TRANSFER","BALANCE_CHECK","DEPOSIT"
  ]

rdd = sc.parallelize(transactions)
```
