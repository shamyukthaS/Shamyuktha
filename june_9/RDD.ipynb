```python
numbers=[10,20,30,40,50]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.map(
    lambda x:x*10
)
print(result.collect())
```

```python
numbers=[1,2,3,4,5]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.map(
    lambda x:x*x
)
print(result.collect())
```

```python
numbers=[1,2,3,4,5,6,7,8]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.filter(
  lambda x:x%2==0
)
print(result.collect())
```

```python
numbers=[10,20,20,30,40,40,50]
rdd=spark.sparkContext.parallelize(numbers)
result=rdd.distinct()
print(result.collect())
```

```python
rdd1=spark.sparkContext.parallelize(
    [10,20,30]
)
rdd2=spark.sparkContext.parallelize(
    [40,50,60]
)
result=rdd1.union(rdd2)
print(result.collect())
```

```python
total_records = rdd.count()
unique_activities = rdd.distinct().count()

most_common = max(
    rdd.countByValue().items(),
        key=lambda x:x[1]
        )
least_common = min(
      rdd.countByValue().items(),
      key=lambda x:x[1]
)
print("===== Learning Platform Usage Summary =====")
print("Total Activities :", total_records)
print("Unique Activities :", unique_activities)
print("Most Common Activity :", most_common)
print("Least Common Activity :", least_common)
```
