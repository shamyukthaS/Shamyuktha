```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("OrdersDataFrame") \
        .getOrCreate()
```

```python
columns = [
      "order_id",
          "customer_name",
              "city",
                  "category",
                      "product",
                          "quantity",
                              "price",
                                  "order_status"
                                  ]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col

df_total = df.withColumn(
    "total_amount",
        col("quantity") * col("price")
        )
df_total.show()
```

```python
from pyspark.sql.functions import lit

df_discount = df_total.withColumn(
    "discount",
        col("total_amount") * lit(0.10)
        )

df_discount.show()
```

```python
df_final = df_discount.withColumn(
      "final_amount",
          col("total_amount") - col("discount")
          )

df_final.show()
```

```python
from pyspark.sql.functions import when

df_category = df_total.withColumn(
    "order_value_category",
        when(col("total_amount") >= 50000,"High")
            .when(col("total_amount") >= 20000,"Medium")
                .otherwise("Low")
                )

df_category.show()
```

```python
df_category.groupBy(
      "order_value_category"
      ).count().show()
```

```python
final_report = df.withColumn(
      "total_amount",
          col("quantity") * col("price")
          ).withColumn(
              "discount",
                  col("quantity") * col("price") * 0.10
                  ).withColumn(
                      "final_amount",
                          (col("quantity") * col("price")) -
                              ((col("quantity") * col("price")) * 0.10)
                              )

final_report.select(
                "order_id",
              "customer_name",
              "city",
              "category",
              "product",
              "quantity",
              "price",
              "total_amount",
              "discount",
              "final_amount",
              "order_status"
            ).show()
```
