```python
import pandas as pd
attendance = pd.read_csv("attendance.csv")
tasks = pd.read_csv("tasks.csv")
print(attendance.head())
print(tasks.head())
```

```python
print(attendance.isnull().sum())
print(tasks.isnull().sum())
```

```python
attendance.fillna("unknown", inplace=True)
tasks.fillna(0, inplace=True)
print("data cleaned successfully")
```

```python
attendance["clock_in"] = pd.to_datetime(attendance["clock_in"])
attendance["clock_out"] = pd.to_datetime(attendance["clock_out"])
attendance["work_hours"] = (
    attendance["clock_out"] -
        attendance["clock_in"]
        ).dt.total_seconds() / 3600
print(attendance.head())
```

```python
attendance["effective_work_hours"] = attendance["work_hours"] - 1
print(attendance.head())
```

```python
attendance["effective_work_hours"] = attendance["work_hours"] - 1
print(attendance.head())
```

```python
task_summary = tasks.groupby(
      "employee_id"
      )["tasks_completed"].sum().reset_index()
merged = pd.merge(
          attendance,
          task_summary,
          on="employee_id"
)
print(merged.head())
```

```python
import pandas as pd
import numpy as np
```

```python
merged["productivity_score"] = np.round(
      merged["tasks_completed"] /
      merged["effective_work_hours"],2
)
print(merged.head())
```

```python
top = merged.sort_values(
      by="productivity_score",
      ascending=False)
print("TOP PERFORMERS")
print(top[[
      "employee_id",
      "productivity_score"]].head())
```

```python
bottom = merged.sort_values(
      by="productivity_score"
      )

print("BOTTOM PERFORMERS")
print(bottom[[
          "employee_id",
          "productivity_score"]].head())
```

```python
absent = merged[
      merged["effective_work_hours"] < 7]
print(absent)
```

```python
report = merged.groupby(
      "employee_id"
      )[[
          "effective_work_hours",
              "productivity_score"]].mean()
print(report)
```

```python
attendance.to_csv(
      "cleaned_attendance.csv",
        index=False)
tasks.to_csv(
      "cleaned_tasks.csv",
      index=False)
report.to_csv("employee_report.csv")
print("files exported successfully")
```
