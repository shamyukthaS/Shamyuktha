```python
accounts = [
    (101,"Chennai",250000),
    (102,"Madurai",180000),
    (103,"Chennai",300000),
    (104,"Coimbatore",150000)
]

columns = ["account_id","city","balance"]

bank_df = spark.createDataFrame(accounts, columns)

display(bank_df)

bank_df.createOrReplaceTempView("bank_accounts")
```

```python
%sql
SELECT city,
SUM(balance) AS total_balance
FROM bank_accounts
GROUP BY city
```
