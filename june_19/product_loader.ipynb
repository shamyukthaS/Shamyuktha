```python
products = [
    (1,"Laptop",50000),
    (2,"Mobile",25000),
    (3,"Tablet",18000),
    (4,"Smart Watch",12000)
]
columns = ["product_id","product_name","price"]
product_df = spark.createDataFrame(products, columns)
display(product_df)
product_df.createOrReplaceTempView("products")
```

```python
%sql
SELECT AVG(price) AS average_price
FROM products
```
