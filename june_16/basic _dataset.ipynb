```python
source_df = spark.createDataFrame(sales_data, columns)
display(source_df)

# 92. Create calculated columns
etl_df = source_df \
    .withColumn("total_amount", col("quantity") * col("price")) \
    .withColumn("tax", col("total_amount") * 0.05) \
    .withColumn("final_amount", col("total_amount") + col("tax")) \
    .withColumn("discount", col("total_amount") * 0.10) \
    .withColumn("net_amount", col("total_amount") - col("discount"))

display(etl_df)
```

```python
completed_df = etl_df.filter(col("status") == "Completed")
display(completed_df)
```

```python
revenue_df = completed_df.withColumn("revenue", col("total_amount"))
display(revenue_df.select("order_id", "product", "city", "category", "revenue"))
```

```python
city_revenue = revenue_df.groupBy("city") \
    .agg(
        count("order_id").alias("total_orders"),
        sum("revenue").alias("total_revenue"),
        avg("revenue").alias("avg_revenue")
    ).orderBy(desc("total_revenue"))
display(city_revenue)
```

```python
category_revenue = revenue_df.groupBy("category") \
    .agg(
        count("order_id").alias("total_orders"),
        sum("revenue").alias("total_revenue"),
        avg("revenue").alias("avg_revenue")
    ).orderBy(desc("total_revenue"))
display(category_revenue)
```

```python
etl_df.write.mode("overwrite").parquet("/tmp/sales_output")
print("Saved as Parquet at /tmp/sales_output")
```

```python
parquet_df = spark.read.parquet("/tmp/sales_output")
display(parquet_df)
```

```python
parquet_df.createOrReplaceTempView("sales_parquet")
display(spark.sql("SELECT * FROM sales_parquet LIMIT 5"))
```

```python
final_report = spark.sql("""
    SELECT
        category,
        city,
        COUNT(order_id)             AS total_orders,
        SUM(quantity)               AS total_units_sold,
        SUM(total_amount)           AS gross_revenue,
        SUM(discount)               AS total_discounts,
        SUM(net_amount)             AS net_revenue,
        ROUND(AVG(total_amount), 2) AS avg_order_value
    FROM sales_parquet
    WHERE status = 'Completed'
    GROUP BY category, city
    ORDER BY gross_revenue DESC
""")
display(final_report)
```
