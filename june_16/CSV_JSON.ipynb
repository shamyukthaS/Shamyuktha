```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

revenue_df = doctors_df.join(
    visits_df,
    "doctor_id",
    "inner"
).groupBy(
    "doctor_id",
    "doctor_name",
    "specialization",
    "city"
).agg(
    F.sum("bill_amount").alias("revenue")
)

revenue_df.show()
```

```python
import warnings
warnings.filterwarnings("ignore", message=".*No Partition Defined for Window.*")
```

```python
window_spec = Window.orderBy(F.col("revenue").desc())

revenue_df.withColumn(
    "rank",
    F.rank().over(window_spec)
).show()
```

```python
revenue_df.withColumn(
    "dense_rank",
    F.dense_rank().over(window_spec)
).show()
```

```python
revenue_df.withColumn(
    "row_number",
    F.row_number().over(window_spec)
).show()
```

```python
revenue_df.withColumn(
    "rank",
    F.rank().over(window_spec)
).filter(
    F.col("rank") == 1
).show()
```

```python
revenue_df.withColumn(
    "rank",
    F.rank().over(window_spec)
).filter(
    F.col("rank") <= 3
).show()
```

```python
spec_window = Window.partitionBy(
    "specialization"
).orderBy(
    F.col("revenue").desc()
)

revenue_df.withColumn(
    "rank",
    F.rank().over(spec_window)
).filter(
    F.col("rank") == 1
).show()
```

```python
revenue_df.withColumn(
    "rank",
    F.rank().over(spec_window)
).filter(
    F.col("rank") <= 2
).show()
```

```python
running_window = Window.orderBy(
    F.col("revenue")
)

revenue_df.withColumn(
    "running_revenue",
    F.sum("revenue").over(running_window)
).show()
```

```python
revenue_df.withColumn(
    "previous_revenue",
    F.lag("revenue",1).over(window_spec)
).show()
```

```python
revenue_df.withColumn(
    "next_revenue",
    F.lead("revenue",1).over(window_spec)
).show()
```

```python
revenue_df.withColumn(
    "previous_revenue",
    F.lag("revenue").over(window_spec)
).withColumn(
    "difference",
    F.col("revenue") - F.col("previous_revenue")
).show()
```

```python
revenue_df.withColumn(
    "next_revenue",
    F.lead("revenue").over(window_spec)
).withColumn(
    "difference",
    F.col("revenue") - F.col("next_revenue")
).show()
```

```python
city_window = Window.partitionBy(
    "city"
).orderBy(
    F.col("revenue").desc()
)

revenue_df.withColumn(
    "rank",
    F.rank().over(city_window)
).filter(
    F.col("rank")==1
).show()
```

```python
city_window_low = Window.partitionBy(
    "city"
).orderBy(
    F.col("revenue").asc()
)

revenue_df.withColumn(
    "rank",
    F.rank().over(city_window_low)
).filter(
    F.col("rank")==1
).show()
```

```python
leaderboard_df = revenue_df.withColumn(
    "rank",
    F.dense_rank().over(window_spec)
).orderBy("rank")

leaderboard_df.show(truncate=False)
```
