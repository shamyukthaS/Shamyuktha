```python
filled_df = bookings_df.na.fill(
      {
              "city":"Unknown",
                      "provider":"Not Available",
                              "booking_status":"Unknown",
                                      "payment_mode":"Not Provided",
                                              "booking_amount":0
                                                  }
                                                  )

filled_df.show()
```

```python
quality_df = bookings_df.withColumn(
      "data_quality_status",
          when(
              col("city").isNull() |
              col("provider").isNull() |
              col("booking_amount").isNull() |
              col("booking_status").isNull(),
                "Incomplete"
                ).otherwise("Complete")
)

quality_df.show()
```

```python
quality_df.groupBy(
      "data_quality_status"
      ).count().show()
```

```python
quality_df.filter(
      col("data_quality_status")=="Complete"
      ).show()
```

```python
quality_df.filter(
      col("data_quality_status")=="Incomplete"
      ).show()
```

```python
tax_df = filled_df.withColumn(
      "tax",
          col("booking_amount") * 0.05
          )

tax_df.show()
```

```python
final_df = tax_df.withColumn(
      "final_amount",
          col("booking_amount") + col("tax")
          )

final_df.show()
```

```python
filled_df.filter(
      col("booking_status")=="Confirmed"
      ).agg(
          sum("booking_amount").alias("Revenue")
          ).show()
```

```python
filled_df.groupBy(
      "service_type"
      ).count().show()
```

```python
filled_df.groupBy(
      "city"
      ).count().show()
```

```python
filled_df.agg(
      avg("booking_amount").alias("Average")
      ).show()
```

```python
bookings_df.dropna(
      subset=["booking_amount"]
      ).agg(
          avg("booking_amount").alias("Average")
          ).show()
```

```python
avg_fill = filled_df.agg(
      avg("booking_amount")
      ).collect()[0][0]

avg_drop = bookings_df.dropna(
      subset=["booking_amount"]
).agg(
        avg("booking_amount")
        ).collect()[0][0]

print("Average after fill:", avg_fill)
print("Average after drop:", avg_drop)
```

```python
filled_df.write.mode(
      "overwrite"
      ).parquet(
          "clean_bookings.parquet"
          )
```

```python
from google.colab import files
uploaded = files.upload()

customers_df = spark.read.option(
    "multiline",
        "true"
        ).json("customers.json")

customers_df.show(truncate=False)
customers_df.printSchema()
```

```python
from pyspark.sql.functions import col

flat_customers_df = customers_df.select(
    "customer_id",
        "name",
            "city",
                "membership",
                    col("contact.phone").alias("phone"),
                        col("contact.email").alias("email"),
                            col("preferences.preferred_service").alias("preferred_service"),
                                col("preferences.budget_range").alias("budget_range")
                                )

flat_customers_df.show(truncate=False)
```

```python
flat_customers_df.filter(
      col("city").isNull()
      ).show()
```

```python
flat_customers_df.filter(
      col("phone").isNull()
      ).show()
```

```python
flat_customers_df.filter(
      col("email").isNull()
      ).show()
```

```python
flat_customers_df.filter(
      col("membership").isNull()
      ).show()
```

```python
flat_customers_df.filter(
      col("preferred_service").isNull()
      ).show()
```

```python
flat_customers_df.filter(
      col("budget_range").isNull()
      ).show()
```

```python
from pyspark.sql.functions import *

flat_customers_df.select([
    count(
            when(col(c).isNull(), c)
                ).alias(c)
                    for c in flat_customers_df.columns
                    ]).show()
```

```python
filled_customers_df = flat_customers_df.na.fill({
      "city":"Unknown",
          "membership":"Standard",
              "phone":"Not Provided",
                  "email":"Not Provided",
                      "preferred_service":"Not Selected",
                          "budget_range":"Unknown"
                          })

filled_customers_df.show()
```

```python
quality_df = flat_customers_df.withColumn(
      "customer_quality_status",
          when(
                  col("city").isNull() |
                          col("phone").isNull() |
                                  col("email").isNull() |
                                          col("membership").isNull() |
                                                  col("preferred_service").isNull(),
                                                          "Incomplete"
                                                              ).otherwise("Complete")
                                                              )

quality_df.show()
```

```python
quality_df.groupBy(
      "customer_quality_status"
      ).count().show()
```

```python
quality_df.filter(
      col("customer_quality_status")=="Complete"
      ).show()
```

```python
quality_df.filter(
      col("customer_quality_status")=="Incomplete"
      ).show()
```

```python
filled_customers_df.groupBy(
      "membership"
      ).count().show()
```

```python
filled_customers_df.groupBy(
      "preferred_service"
      ).count().show()
```

```python
flat_customers_df.write.mode(
      "overwrite"
      ).parquet(
          "customers_flat.parquet"
          )
```

```python
filled_customers_df.write.mode(
      "overwrite"
      ).option(
          "header",
              "true"
              ).csv(
                  "clean_customers.csv"
                  )
```
