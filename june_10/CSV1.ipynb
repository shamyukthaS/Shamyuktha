```python
df_tax = df.withColumn(
      "tax",
          col("ticket_price") * 0.05
          )
df_tax.show()
```

```python
df_final = df.withColumn(
      "tax",
          col("ticket_price") * 0.05
          ).withColumn(
              "final_price",
                  col("ticket_price") + (col("ticket_price") * 0.05)
                  )

df_final.show()
```

```python
df_category = df.withColumn(
      "price_category",
          when(col("ticket_price") >= 30000,"Premium")
              .when(col("ticket_price") >= 10000,"Standard")
                  .otherwise("Budget")
                  )

df_category.show()
```

```python
df_category.groupBy("price_category").count().show()
```

```python
df.filter(col("status")=="Confirmed") \
  .write.mode("overwrite") \
    .parquet("confirmed_flights.parquet")
```
